## Overview of Projects

Throughout the academic year, I contributed to and led development on two major full-stack projects: **Neptune** and **CrossWise**. Each project pushed my capabilities as a developer and demonstrated my leadership, adaptability, and commitment to real-world problem-solving.

### **Project: Neptune**

**GitHub:** [Frontend](https://github.com/DNHS-Neptune/neptune_frontend) | [Backend](https://github.com/DNHS-Neptune/neptune_backend)

**Summary:** Neptune is a reimagined version of the Saturn dashboard used by our school. It provides theme personalization, profile features, and student-focused design improvements.

**My Contributions:**

* Developed and deployed a dynamic **theme selector** with full **CRUD operations**.
* Extended backend schema with `themes` and `users` tables, linking user profiles to custom CSS.
* Implemented real-time frontend updates using asynchronous JavaScript calls to Python Flask APIs.
* Managed **AWS deployment** and troubleshooting issues during rollout.
* Presented the full system during **Night at the Museum (Trimester 2)**.

**Skills Gained:**

* CRUD with Python
* Asynchronous JS + REST API integration
* AWS EC2 deployment
* UI/UX logic and real-time theming
* Git collaboration and project structure

### **Project: CrossWise**

**GitHub:** [Frontend](https://github.com/illuminati1618/CrossWise) | [Backend](https://github.com/illuminati1618/CrossWise_backend)

**Summary:** CrossWise is a planning assistant for cross-border travel, offering **real-time wait time predictions** and **NOAA-based weather forecasts**.

**My Contributions:**

* Designed and implemented initial **machine learning model** for wait time predictions.
* Pivoted from ML to NOAA data integration to improve **accuracy and reliability**.
* Built full-stack solution with Flask, JavaScript, Chart.js, and NOAA APIs.
* Ensured frontend responsiveness and backend data caching.
* Presented live demo at **Night at the Museum (Trimester 3)**.

**Skills Gained:**

* Machine learning prototyping and dataset analysis
* API integration (NOAA)
* Data visualization with Chart.js
* Independent research and technical decision-making
