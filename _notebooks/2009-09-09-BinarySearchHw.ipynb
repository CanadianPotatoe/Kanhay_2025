# Decimal to Binary Conversion

Convert the following decimal numbers into binary, and then explain the place values used:

## 1. Convert **42** to Binary

### Division-by-2 Steps

| Step | Number | Divided by 2 | Quotient | Remainder |
|:----:|:------:|:------------:|:--------:|:---------:|
| 1 | 42 | 42 ÷ 2 | 21 | 0 |
| 2 | 21 | 21 ÷ 2 | 10 | 1 |
| 3 | 10 | 10 ÷ 2 | 5 | 0 |
| 4 | 5 | 5 ÷ 2 | 2 | 1 |
| 5 | 2 | 2 ÷ 2 | 1 | 0 |
| 6 | 1 | 1 ÷ 2 | 0 | 1 |

### Binary Result
Read the remainders **bottom to top**:  
**42 in binary = `101010`**

### Place Value Table

| 32 | 16 | 8 | 4 | 2 | 1 |
|----|----|---|---|---|---|
| 1  | 0  | 1 | 0 | 1 | 0 |

**Explanation:**  
\( 32 + 8 + 2 = 42 \)

# Free Response Questions

### 1. Why is binary search more efficient for large datasets than linear search?
Binary search is faster because it cuts the list in half each time. Linear search checks each item one by one.

---

### 2. What happens if the list is not sorted and binary search is used?
It won’t work right. Binary search needs the list to be sorted to know where to look next.

---

### 3. Could binary search be used in a video game leaderboard or a streaming service search bar? Why or why not?
Yes, if the list is sorted. But usually they use better search tools that work for more types of input.
